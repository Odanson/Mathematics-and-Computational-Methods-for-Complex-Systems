```python
def my_transf(B):
    rows = len(B)
    cols = len(B[0])
    B_transpose = [[0] * rows for _ in range(cols)]

    for i in range(rows):
        for j in range(cols):
            B_transpose[j][i] = B[i][j]

    result = [[B[i][j] + B_transpose[i][j] for j in range(cols)] for i in range(rows)]

    return np.array(result)
```

Let's try our function to ensure it works well

```python
matrix_B = np.array([[2, 1, 0],
                    [5, 4, 3],
                    [9, 7, 6]])


re_sult = my_transf(matrix_B)


print("Matrix B is given by:")
print(matrix_B)
print()
print("Sum of B and Bᵀ:")
re_sult
```

### 1.b.

We will implement the eigenvalues and eigenvectors using the `np.linalg.eig` function from `NumPy`. This enables us to compute the eigenvectors and eigenvalues of matrix B with ease. Also, the NumPy library affords us the `np.dot` function to compute the dot products for the $i^{th}$ and $j^{th}$ eigenvectors which comes at no extra computational cost.

```python
def eigprod(B, i, j):
    eigenvalues, eigenvectors = np.linalg.eig(B)

    # Getting the i-th and j-th eigenvectors
    v_i = eigenvectors[:, i]
    v_j = eigenvectors[:, j]

    # Now we calculate the dot product of the eigenvectors
    dot_product = np.dot(v_i, v_j)

    return dot_product
```

### Observation

We know that the dot product of two vectors $\vec{v}$ and $\vec{w}$ is defined as:

$\vec{v} \cdot \vec{w} = \sum_{i=1}^{n} v_i w_i$. 

Therefore, for any two vectors $v_i$ and $v_j$, the dot product is given by:

 $$v_i \cdot v_j = \sum_{k=1}^{n} v_{i,k} \cdot v_{j,k}$$
 
Where, $v_{i,k}$ and $v_{j,k}$ are the components of vectors $v_i$ and $v_j$ in the $\text {k-th}$ dimension. The dot product measures the degree of alignment between the vectors.

#### The Orthogonality Property of Eigenvectors

If $A$ is a square matrix with distinct eigenvalues $\lambda_1$, $\lambda_2$, $\ldots$, $\lambda_n$, and $v_1$, $v_2$, $\ldots$, $v_n$ are the corresponding eigenvectors, then the eigenvectors are orthogonal.

This means that:

$v_i^T \times v_j = 0 \quad \text{for} \ i \neq j$

Here:
- $v_i^T$ is the transpose of the $\text{i-th}$ eigenvector.
- $v_j$ is the $\text{j-th}$ eigenvector.


In other words, the dot product of distinct eigenvectors is zero, indicating that they are orthogonal to each other. This property holds because eigenvectors corresponding to distinct eigenvalues are associated with different directions in the vector space, and orthogonality reflects the independence of these directions.

**NOTE:** It's important to note that this property assumes distinct eigenvalues. In cases where eigenvalues are repeated (have multiplicity greater than 1), the corresponding eigenvectors might not be orthogonal.

The observation from our result is consistent with the properties of eigenvectors and their orthogonality as explained above.

- When $i = j$, we computed the dot product of an eigenvector with itself. This dot product should be close to 1, indicating that the eigenvector is normalised (has a length of 1).

- When $i \neq j$, we computed the dot product of different eigenvectors. Since eigenvectors corresponding to distinct eigenvalues are orthogonal, the dot product should be close to 0.

So, the observation aligns with these expectations: diagonal elements close to 1 (indicating normalized eigenvectors) and off-diagonal elements close to 0 (indicating orthogonality between different eigenvectors). Numerical precision may cause small deviations from exact 0 or 1 values.

### 1.e.

Due to the orthogonality of eigenvectors corresponding to distinct eigenvalues in symmetric matrices, the eigenvectors $v_2(R)$ and $v_3(R)$ are linearly independent.

#### Linear Independency

Two vectors $\vec{a_1}$ and $\vec{a_2}$ are linearly independent if the only way to satisfy the equation $c_1 \cdot \vec{a_1} + c_2 \cdot \vec{a_2} = 0$ is if $c_1 = c_2 = 0$, where $c_1$ and $c_2$ are scalars.

Since the vectors here, say $v_1$ and $v_2$, are eigenvectors of a symmetric matrix $R$ with distinct eigenvalues $(\lambda_1 \neq \lambda_2)$, we know that $v_1^T \cdot v_2 = 0$ due to orthogonality.



Consider the linear combination $c_1 \cdot v_1 + c_2 \cdot v_2 = 0$. Take the dot product of both sides with $v_2$:

$(c_1 \cdot v_1 + c_2 \cdot v_2)^T \cdot v_2 = 0 $

Expanding and using properties of transposition:

$ (c_1 \cdot v_1^T + c_2 \cdot v_2^T) \cdot v_2 = 0 $

$ c_1 \cdot v_1^T \cdot v_2 + c_2 \cdot v_2^T \cdot v_2 = 0 $

Since $v_1^T \cdot v_2 = 0$ (due to orthogonality), the term $c_1 \cdot v_1^T \cdot v_2$ becomes zero, and we are left with:

$ c_2 \cdot v_2^T \cdot v_2 = 0 $

Now, because $v_2$ is an eigenvector, it is nonzero, and $v_2^T \cdot v_2$ is the square of the magnitude of $v_2$. The magnitude of $v_2$ is nonzero, so $c_2$ must be zero to satisfy the equation.

Now, let's repeat the process by taking the dot product with $v_1$:

$ (c_1 \cdot v_1 + c_2 \cdot v_2)^T \cdot v_1 = 0 $

$ c_1 \cdot v_1^T \cdot v_1 + c_2 \cdot v_2^T \cdot v_1 = 0 $

Since $v_1^T \cdot v_1$ is the square of the magnitude of $v_1$, and $v_1$ is an eigenvector (nonzero), $c_1$ must be zero to satisfy the equation.

Therefore, the only way for $c_1 \cdot v_1 + c_2 \cdot v_2 = 0$ to hold is if $c_1 = c_2 = 0$, which means that $v_1$ and $v_2$ are linearly independent.

Therefore the eigenvectors $v_2(R)$ and $v_3(R)$, and by extension any other eigenvector of a symmetric matrix, are not linearly dependent.

### 1.f

The linear independence of eigenvectors of a symmetric matrix implies that these eigenvectors span a subspace of the full vector space, and the dimension of this subspace is equal to the number of distinct eigenvalues, providing important insights into the structure and diagonalisability of the matrix.

When eigenvectors corresponding to distinct eigenvalues are linearly independent, it means that each eigenvector contributes a unique direction to the vector space. In other words, the set of linearly independent eigenvectors spans a subspace of the full vector space, and the dimension of this subspace is equal to the number of linearly independent eigenvectors.

For a symmetric matrix, the spectral theorem guarantees that the eigenvectors corresponding to distinct eigenvalues are orthogonal (and hence linearly independent). Therefore, if the symmetric matrix $R$ has $n$ distinct eigenvalues, the dimension of the vector space spanned by the eigenvectors is $n$. 


**Claim:** If $R$ is a symmetric matrix with $n$ distinct eigenvalues $\lambda_1, \lambda_2, \ldots, \lambda_n$ and corresponding eigenvectors $v_1, v_2, \ldots, v_n$, then the dimension of the vector space spanned by these eigenvectors is $n$.

This can be proven from the orthogonality and linear independence of eigenvectors corresponding to distinct eigenvalues, which is a consequence of the spectral theorem for symmetric matrices as shown below:

1. **Linear Independence:**
   - By the spectral theorem for symmetric matrices, eigenvectors corresponding to distinct eigenvalues are orthogonal (and hence linearly independent).
   - Let $v_1, v_2, \ldots, v_n$ be eigenvectors corresponding to distinct eigenvalues $\lambda_1, \lambda_2, \ldots, \lambda_n$. Due to orthogonality, these vectors are linearly independent.

2. **Spanning the Vector Space:**
   - Every eigenvector corresponds to a unique eigenvalue, and the set $\{v_1, v_2, \ldots, v_n\}$ spans a subspace of the full vector space.
   - Since the eigenvectors are linearly independent, they form a basis for this subspace.

3. **Dimension of the Subspace:**
   - The dimension of the subspace spanned by the eigenvectors is equal to the number of linearly independent eigenvectors. In this case, it is $n$.

Therefore, the dimension of the vector space spanned by the eigenvectors of $R$ is $n$, where $n$ is the number of distinct eigenvalues.


#### Significance:
This is particularly significant because, in the context of diagonalization, these linearly independent eigenvectors can form a basis for the vector space. In practical terms, it means that the symmetric matrix $R$ can be diagonalized by a matrix $P$ composed of its linearly independent eigenvectors, leading to the diagonal matrix $\Lambda$ with the eigenvalues on the diagonal: $R = P \Lambda P^{-1}$.


Mathematically, these concepts can be summarized as follows:

$ Rv_i = \lambda_i v_i \text{ for } i = 1, 2, \ldots, n $

$ \text{Linear Independence: } \sum_{i=1}^{n} c_i \cdot v_i = 0 \implies c_1 = c_2 = \ldots = c_n = 0 $

$ \text{Orthogonality: } v_i^T \cdot v_j = 0 \text{ for } i \neq j $

$ \text{Vector Space: } \{v_1, v_2, \ldots, v_n\} \text{ spans a subspace } V $

$ \text{Dimension of Subspace: } \dim(V) = n $

$ \text{Spectral Theorem: For a symmetric matrix } R, \text{ eigenvectors are orthogonal.} $

$ \text{Diagonalization: } R = P \Lambda P^{-1} \text{ where } P = [v_1, v_2, \ldots, v_n] \text{ and } \Lambda = \text{diag}(\lambda_1, \lambda_2, \ldots, \lambda_n) $

### Q.2.a.

The `remain_prob()` function can be constructed using the cumulative probability of surviving the initial 100 months without a fault. This cumulative probability is the complement of the cumulative failure probability.

```python
def remain_prob():
    total_probability = 1.0  

    for t in range(1, 101):
        overuse_fault_probability = 0.0001 * t**2 / (1 + t)
        factory_fault_probability = 0.01 * (1 + (1 - t) / (1 + t))

        survival_prob_t = 1 - (overuse_fault_probability + factory_fault_probability)
        total_probability *= survival_prob_t

    return total_probability
```

### Running `remain_prob` Function

```python
remaining_probability = remain_prob()
print(f"The probability of a CPT surviving the first 100 months without a fault is: {remaining_probability:.4f}")
```

### Q.2.b.

```python
def factory_simulation(n0):
    v = np.zeros(100, dtype=int)
    
    for month in range(1, 101):
        for i in range(n0):
            # Simulate overuse and factory faults
            overuse_fault_prob = 0.0001 * month**2 / (1 + month)
            factory_fault_prob = 0.01 * (1 + (1 - month) / (1 + month))
            
            # Check if a fault occurs
            if np.random.rand() < (overuse_fault_prob + factory_fault_prob):
                n0 -= 1  
        
        v[month - 1] = n0  
    
    return v

# Simulate faults with an initial number of 100 CPTs
initial_CPTs = 100
simulation_result = factory_simulation(initial_CPTs)

# Plot the output
plt.scatter(range(1, 101), simulation_result)
plt.xlabel("Months")
plt.ylabel("Functioning CPTs")
plt.ylim(0, initial_CPTs)
plt.title("Factory Simulation")
plt.grid(True)
plt.show()
```

### Q.2.c.

```python
def profit(x, y):
    total_profit = 0.0

    for t in range(1, 101):
        k_x = max(1 - 20 * x, 0)
        k_y = max(1 - 20 * y, 0)

        overuse_fault_prob = k_x * 0.0001 * t**2 / (1 + t)
        factory_fault_prob = k_y * 0.01 * (1 + (1 - t) / (1 + t))

        item_profit = 5.0 * (1 - (overuse_fault_prob + factory_fault_prob))

        total_profit += item_profit

    expected_profit = 100 * total_profit

    return expected_profit


investment_x = 0.02
investment_y = 0.01

expected_profit_result = profit(investment_x, investment_y)
print(f"Expected profit for 100 items: £{expected_profit_result:.2f}")
```

### Q.2.d.(ii)

The condition for locally optimal allocation of investment to occur is when the gradient (or Jacobian in the case of vector-valued functions) of the objective function is equal to zero. This is based on the first-order optimality condition.

Mathematically, for a function $p(x, y)$, the first-order optimality condition for a local optimum requires that the gradient vector is equal to the zero vector:

$$
\nabla p(x, y) = \mathbf{0}
$$

Here, $\nabla$ represents the gradient operator, and $\mathbf{0}$ is the zero vector. The gradient vector contains the partial derivatives of the function with respect to each variable $x$ and $y$.

This condition ensures that, at the point of a local optimum, the rate of change of the profit function with respect to each investment variable is zero. This means that moving slightly in any direction from the optimal point should not result in an increase in profit.

To find the local optimum, you would typically solve the system of equations formed by setting the partial derivatives of the objective function equal to zero:

$$
\frac{\partial p}{\partial x} = 0 \quad \text{and} \quad \frac{\partial p}{\partial y} = 0
$$

This system represents the critical points where the gradient is zero and can be solved to find the optimal values of $x$ and $y$. 

- **NOTE:** This condition identifies only potential local optima. Additional analysis may be needed to confirm the nature of the points (whether they are minima, maxima, or saddle points).

The sample space describes all possible outcomes of an experiment. Here, there are two uncertain events:

To describe the sample space, we can list all possible pairs of events $(B_C, B_R)$:

- $(B_1, B_1)$: You choose Box 1, and the money is in Box 1.
- $(B_1, B_2)$: You choose Box 1, but the money is in Box 2.
- $(B_1, B_3)$: You choose Box 1, but the money is in Box 3.

- $(B_2, B_1)$: You choose Box 2, and the money is in Box 1.
- $(B_2, B_2)$: You choose Box 2, and the money is in Box 2.

- $(B_3, B_2)$: You choose Box 3, but the money is in Box 2.
- $(B_3, B_3)$: You choose Box 3, and the money is in Box 3.

Each of these pairs represents a possible outcome, and together they form the sample space for this scenario.

Therefore the sample space can be represented as follows:

$$ \Omega = \{(B_1, B_1), (B_1, B_2), (B_1, B_3), (B_2, B_1), (B_2, B_2), (B_2, B_3), (B_3, B_1), (B_3, B_2), (B_3, B_3)\} $$

Here, $ \Omega $ represents the sample space, and each element in the set represents a possible outcome in the form $(B_C, B_R)$, where $B_C$ is the chosen box and $B_R$ is the box containing the money.

### Q.3.b.

To calculate the expected reward $(\text{E}[R])$ and the variance of the reward $(\text{Var}(R))$, we need to consider the probability distribution of the random variable \(R\), which represents the amount of money you can earn.

Let the amount of money in a given box be $X$, and let $p_i$ be the probability of choosing box $B_i$ for $i = 1, 2, 3$. The expected reward is then given by:

$ \text{E}[R] = \sum_{i=1}^3 p_i \cdot X_i $

The variance of the reward is given by:

$ \text{Var}(R) = \sum_{i=1}^3 p_i \cdot (X_i - E[R])^2 $

For uniform probability, each box has the same probability of being chosen, i.e., $p_i = \frac{1}{3}$ for each $i$. Therefore, the formulas simplify to:

$ \text{E}[R] = \frac{1}{3} \sum_{i=1}^3 X_i $

$ \text{Var}(R) = \frac{1}{3} \sum_{i=1}^3 (X_i - \text{E}[R])^2 $

### Q.3.c.

We can approach answering this question by separating the outcomes into two events: Sticking to initial choice and swapping for the final box


Let's denote the box initially chosen as $B_C$ and the box with the money as $B_R$. The reward variable is still denoted as $R$. The probabilities of initially choosing each box are $p_i = \frac{1}{3}$ for $i = 1, 2, 3$.

Considering our two events as suggested:

1. Event $A$: You initially, unknowingly choose the correct box. The probability of this event is $P(A) = \frac{1}{3}$.

2. Event $B$: You choose one of the incorrect boxes. The probability of this event is $P(B) = \frac{2}{3}$.

Now, we can express the expected value and variance based on these events:

### Stick Strategy:
$E[R \,|\, \text{Stick}] = P(A) \cdot X_R + P(B) \cdot 0$
$E[R \,|\, \text{Stick}] = \frac{1}{3} \cdot X_R$

$ \text{Var}(R \,|\, \text{Stick}) = P(A) \cdot (X_R - E[R \,|\, \text{Stick}])^2 + P(B) \cdot (0 - E[R \,|\, \text{Stick}])^2 $
$ \text{Var}(R \,|\, \text{Stick}) = \frac{1}{3} \cdot \left(X_R - \frac{1}{3} \cdot X_R\right)^2 $

### Swap Strategy:
$E[R \,|\, \text{Swap}] = P(A) \cdot 0 + P(B) \cdot X_R$
$E[R \,|\, \text{Swap}] = \frac{2}{3} \cdot X_R$

$ \text{Var}(R \,|\, \text{Swap}) = P(A) \cdot (0 - E[R \,|\, \text{Swap}])^2 + P(B) \cdot (X_R - E[R \,|\, \text{Swap}])^2 $
$ \text{Var}(R \,|\, \text{Swap}) = \frac{2}{3} \cdot \left(X_R - \frac{2}{3} \cdot X_R\right)^2 $

These formulas represent the expected value and variance of the reward under the 'stick' strategy and the 'swap' strategy, considering the events of initially choosing the correct box or one of the incorrect boxes.

```python
def parameterise_basic_model(p):
    a, b, c, d = p

    def model(x, t):
        N, P = x
        return [a * x[0] - b * np.prod(x),
                c * np.prod(x) - d * x[1]]

    return model


basic_model = parameterise_basic_model([1, 4, 1, 4])


def solve(f, tspan, x0):
    δt = 0.01
    ts = np.arange(tspan[0], tspan[1] + δt, δt)
    xs = np.zeros((len(x0), len(ts)))
    xs[:, 0] = x0

    for i in range(1, len(ts)):
        xs[:, i] = xs[:, i-1] + δt * np.array(f(xs[:, i-1], ts[i]))

    return ts, xs.T


t_span = (0, 10)
x_0 = np.array([1.0, 1.0])
ts, xs = solve(basic_model, t_span, x_0)

plt.plot(ts, xs[:, 0], label="prey", linewidth=4)
plt.plot(ts, xs[:, 1], label="predators", linewidth=4)
plt.xlabel("time")
plt.ylabel("population")
plt.legend(fontsize=16)
```

### Q.4.a.

Here, we are looking for a situation where the rate at which new prey are born is equal to the rate at which prey are dying due to predators. This happens when the two rates balance each other out, and the population stays constant.

To do this, we need to find the maximum predator population for which prey births still outstrip deaths, we can set the rate of change of prey population $(\dot{N}(t))$ to zero and solve for the predator population $(P(t))$ at that point. The birth rate of prey is given by $a N(t)$, and the death rate of prey is given by $b N(t) P(t)$.

This means, we set $\dot{N}(t)$ to zero:

$a N(t) - b N(t) P(t) = 0$

Solving the equation for $P(t)$:

$a = b P(t)$

$P(t) = \frac{a}{b}$

Therefore, the maximum predator population $(P(t))$ for which prey births still outstrip deaths is given by:

$P_{\text{max}} = \frac{a}{b}$

This is the critical point where the birth and death rates of prey are equal, and beyond this point, the death rate of prey exceeds the birth rate.

### Q.4.b.

The predator population $(P(t))$ will be shrinking when the rate of change of the predator population $(\dot{P}(t))$ is negative. In the given predator-prey model:

$\dot{P}(t) = c N(t) P(t) - d P(t)$

The term $c N(t) P(t)$ represents the birth rate of predators, and $d P(t)$ represents the death rate of predators. If the death rate of predators is greater than the birth rate, the overall rate of change of the predator population will be negative, indicating a decrease in the predator population.

Therefore, the conditions under which the predator population is shrinking are when:

$$d > c N(t)$$

That is to say, the death rate of predators $(d)$ must be greater than the product of the parameter $(c)$ and the prey population $N(t)$ for the predator population to decrease. This could happen if the prey population is not sufficient to sustain the predator population, leading to a decline in predators.

### Q.4.c(i)

The predator-prey model equations are:

$$\dot{N}(t) = a N(t) - b N(t) P(t)$$
$$\dot{P}(t) = c N(t) P(t) - d P(t)$$


In the absence of predators we have $(P(0) = 0)$, and we only need to consider the dynamics of the prey population $(N(t))$.

Since $(P(0) = 0)$, the second equation becomes:

$\dot{P}(t) = 0$

This means that the predator population does not exist initially (at $t = 0$), and it will stay zero for all time $(\dot{P}(t) = 0)$. Therefore, we can focus only on the dynamics of the prey population.

The first equation becomes:

$\dot{N}(t) = a N(t)$

This is a simple exponential growth equation. The rate of change of the prey population is directly proportional to the current population size, with the growth rate $a$. 

Thus, in the absence of predators, the prey population will grow exponentially over time. The larger the value of $a$, the faster the prey population will increase.

### Q.4.c(ii).

Yes, the new model in the absence of predators $(P(0) = 0)$ is reasonable, and it represents a scenario of prey population growth without the influence of predators. An example is a classic exponential growth model, commonly used to describe population growth in the absence of limiting factors as shown below:

$\dot{N}(t) = a N(t)$

Where:

### Q.4.h.

Using the gradient function provided, we can define a gradient descent function as thus:

```python
def gradient(f, x):
    n = len(x)
    δx = 0.01

    def Δx(i):
        z = np.zeros(n)
        z[i] = δx
        return z

    dfdx = np.zeros(n)
    for i in range(n):
        dfdx[i] = (f(x + Δx(i)) - f(x)) / δx

    return dfdx
```

```python
def gradient_descent(f, initial_params, learning_rate, num_iterations):
    current_params = np.array(initial_params, dtype=float)

    for _ in range(num_iterations):
        grad = gradient(f, current_params)
        current_params -= learning_rate * grad

    return current_params
```

### Q.5.e. The Jacobian matrix $J(x)$

We know that the Jacobian matrix $J(x)$ of the system $\dot{x}(t) = f(x(t))$ w.r.t. the vector $x(t)$ is given by the partial derivatives of the components of $f$ w.r.t. the components of $x$. 

Mathematically, this is expressed as:

$ J(x) = \frac{\partial f}{\partial x}(x) = \begin{bmatrix}
\frac{\partial f_1}{\partial x_1} & \frac{\partial f_1}{\partial x_2} \\
\frac{\partial f_2}{\partial x_1} & \frac{\partial f_2}{\partial x_2}
\end{bmatrix} $

Here, $f(x) = [f_1(x), f_2(x)]$ is the vector function defining the system.

To compute the Jacobian matrix, we need to find the partial derivatives of each component of $f$ w.r.t. each component of $x$.

If $f(x) = [f_1(x), f_2(x)]$, then the Jacobian matrix is:

$ J(x) = \begin{bmatrix}
\frac{\partial f_1}{\partial x_1} & \frac{\partial f_1}{\partial x_2} \\
\frac{\partial f_2}{\partial x_1} & \frac{\partial f_2}{\partial x_2}
\end{bmatrix} $

The elements of this matrix are the partial derivatives:

$ J(x) = \begin{bmatrix}
\frac{\partial f_1}{\partial x_1} & \frac{\partial f_1}{\partial x_2} \\
\frac{\partial f_2}{\partial x_1} & \frac{\partial f_2}{\partial x_2}
\end{bmatrix} $

But we already have our system of differential equations given by:

$\frac{d u_1}{d \tau} = u_1(1 - u_1 - a_{12}u_2)$

$\frac{d u_2}{d \tau} = \rho u_2(1 - u_2 - a_{21}u_1)$



Now, let's calculate the Jacobian matrix $J(x)$:

$ \frac{\partial f_1}{\partial u_1} = 1 - 2u_1 - a_{12}u_2 $



$ \frac{\partial f_1}{\partial u_2} = -a_{12}u_1 $



$ \frac{\partial f_2}{\partial u_1} = -\rho a_{21}u_2 $



$ \frac{\partial f_2}{\partial u_2} = \rho(1 - 2u_2 - a_{21}u_1) $

Substituting these partial derivatives into the Jacobian matrix \(J(x)\):

$$ J(x) =
\begin{bmatrix}
1 - 2u_1 - a_{12}u_2 & -a_{12}u_1 \\
-\rho a_{21}u_2 & \rho(1 - 2u_2 - a_{21}u_1) \\
\end{bmatrix}
$$

This is similar to our previous result. Thus, confirmed.

### Requirements on the Values of $a_{12}$ and $a_{21}$ for Fixed Points Stability

The stability of fixed points in a dynamical system is determined by the eigenvalues of the Jacobian matrix evaluated at those fixed points. For stability, all eigenvalues must have negative real parts.

Let's denote the eigenvalues of the Jacobian matrix $J(x)$ at a fixed point as $\lambda_1$ and $\lambda_2$. The fixed point will be stable if both $\text{Re}(\lambda_1) < 0$ and $\text{Re}(\lambda_2) < 0$.


### Fixed Point $x_1$:

Eigenvalues: $\lambda_1, \lambda_2$

For stability:
- $\text{Re}(\lambda_1) < 0$
- $\text{Re}(\lambda_2) < 0$

### Fixed Point $x_2$:

Eigenvalues: $\lambda_1, \lambda_2$

For stability:
- $\text{Re}(\lambda_1) < 0$
- $\text{Re}(\lambda_2) < 0$



At fixed point $x = [u_1, u_2]$ our Jacobian matrix is given by:

$$ J(x) =
\begin{bmatrix}
1 - 2u_1 - a_{12}u_2 & -a_{12}u_1 \\
-\rho a_{21}u_2 & \rho(1 - 2u_2 - a_{21}u_1) \\
\end{bmatrix}
$$

For stability, the following conditions are implied:

1. $1 - 2u_1 - a_{12}u_2 < 0$ (which is the real part of the first eigenvalue is negative)
2. $\rho(1 - 2u_2 - a_{21}u_1) < 0$ (the real part of the second eigenvalue is negative)



For the first condition to be satisfied:
   $ 1 - 2u_1 - a_{12}u_2 < 0 $
   
   Solving for $a_{12}$:
   
   $ a_{12} > \frac{1 - 2u_1}{u_2} $
   

For the second condition to be satisfied:
   $ \rho(1 - 2u_2 - a_{21}u_1) < 0 $
   
   Solving for $a_{21}$:
   
   $ a_{21} > \frac{1 - 2u_2}{\rho u_1} $

Therefore, the requirements on the values of $a_{12}$ and $a_{21}$ for stability at a fixed point $x = [u_1, u_2]$ are:

$ a_{12} > \frac{1 - 2u_1}{u_2} $

$ a_{21} > \frac{1 - 2u_2}{\rho u_1} $
